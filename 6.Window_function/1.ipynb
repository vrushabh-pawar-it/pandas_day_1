```python
# Window function = calculation on a moving group of rows (not whole column)
```

```python
import pandas as pd
import numpy as np
```

```python
# “Does each row depend on previous rows or a subset of rows?”
# yes means here used a compulsory window function
```

```python
df = pd.DataFrame({
    'a': [1,2,3,4,5,6],
    'b':[10, np.nan, np.nan, np.nan, np.nan, 30]
})
df
```

```python
# cumulative sum
df['a_cumsum'] = df['a'].cumsum()
df
```

```python
# cummu prod of 'a'
df['cumprod_a'] = df['a'].cumprod()
print("the prod is = ", df)
```
