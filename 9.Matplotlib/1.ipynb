```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\91942\Desktop\11.11\agree\SNOW\pandas\datasets1\datasets1\score.csv')
df
```

```python
# line plot
x = df['marks']
y = df['rank']
plt.plot(x,y, color='orange')
plt.xlabel('Student marks', color='red')
plt.ylabel('Student ranks', color = 'blue')
plt.title('Exam marks rank')
plt.show

    
```

```python
# scatter plot
x = df['marks']
y = df['rank']
plt.plot(x,y, color = 'orange')
plt.scatter(x,y, color = 'green')
plt.xlabel('Student marks', color='red')
plt.ylabel('Student rank', color='blue')
plt.title('exam marks rank', color = 'pink')
plt.show
```

```python
# bar graph
x = df['marks']
y = df['rank']
plt.bar(x,y,color='orange')
plt.xlabel('student marks', color = 'red')
plt.ylabel('Student rank', color = 'blue')
plt.title('Exam marks rank', color = 'cyan')
plt.show
```

```python
#  how many students got above 850 marks
x1 = len(x[x>850])
x1
```

```python
# how many got marks between 800 and 900
x2 = len(x[(x>800) & (x<900)])
x2
```

```python
# how many student got marks less than 800
x3 = len(x[x<800])
print("the student less than 800 mraks is :", x3)
```

```python
print(x1,x2,x3)
```

```python
slices = [x1,x2,x3]
cols = ['yellow','red','green']
lbls = ['>900', '800 to 900', '<800']

plt.pie(
    slices,
    labels = lbls,
    colors = cols,
    startangle = 90,
    shadow = True,
    explode=(0.2,0,0)
)
plt.show()
```

```python
# sub plot
fig = plt.figure(figsize = (12,7)) #height and width of figure

#1
g1 = fig.add_subplot(2,2,1, facecolor = 'yellow') # row column and postition means left side on top position
g1.plot(x,y)
g1.set_title('mraks va rank')

#2
g2 = fig.add_subplot(2,2,2, facecolor = 'green')
g2.plot(x,y)
g2.set_title('rank vs admission')

#3
mx = max(x)
mn = min(y)
g3 = fig.add_subplot(2,2,3, facecolor = 'green')
g3.pie(
    [mx,mn],
    colors=['yellow','red'],
    labels = ['max marks', 'min marks'],
    startangle = 90
)
g3.set_title('mark-rank-admission')
plt.tight_layout()
plt.show()
```
