```python
# groupby() method -> to split the data into groups and then apply any operation on groups
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\91942\Downloads\datasets1\datasets1\cities_weather.csv')
df
```

```python
df1 = df.groupby('city')
df1
```

```python
for city, data in df1:
    print(data)
```

```python
# Aggregate function : are the function that act on 
# group of element max(), min(), mean(),
# sum(), median(), std(), prod(), count()
```

```python
df1.get_group('mumbai')
```

```python
df1.get_group('paris')
```

```python
df1.get_group('mumbai').max()
```

```python
df1.agg('max')
```

```python
df1.agg('min')
```

```python
df1['temperature'].agg(['max', 'min'])
```

```python
# find avg speed in New york
df1 = df.groupby('city')['windspeed'].mean()
df1['new york']
print("the avg windspeed new york is = ", df1['new york'])
```

```python
# min temp and max windspeed in new york
temp = df.groupby('city')['temperature'].min()
wind = df.groupby('city')['windspeed'].max()
print("the minimum tem is = ", temp['new york'])
print("the max windspeed is =", wind['new york'])
```

```python
# 2nd way
df1.agg({'temperature': 'min', 'windspeed': 'max'})
```

```python
#  find min, max on temperature - sum and avg on windspeed ?
df = df.groupby('city')
df1 = df.get_group('new york')
df1.agg({
    'temperature': ['min', 'max'],
    'windspeed': ['sum', 'mean']
})
```

```python
# total dataset
df.agg({'temperature' : ['min', 'max'], 'windspeed' : ['sum', 'mean']})
```
