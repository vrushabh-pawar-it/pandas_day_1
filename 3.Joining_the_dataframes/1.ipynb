```python
import pandas as pd
```

```python
df1 = {'player name': ['rahul dravid', 'virat kohli', 'rohit sahrma','ricky ponting'],
       'age': [10,20, 40, 30]}
df1
```

```python
df1 = pd.DataFrame({'player name': ['rahul dravid', 'virat kohli', 'rohit sahrma','ricky ponting'],
       'age': [10,20, 40, 30]})
df1
```

```python
df2 = pd.DataFrame({
    'player name' : ['rahul dravid', 'sachin tendulkar', 'virat kohli', 'Hardik Pandya', 'M.S. Dhoni'],
    'salary' : [1000000, 500000, 2000000, 1500000, 2500000]
})

df2
```

```python
# inner join
pd.merge (df1, df2, on = 'player name', how = 'inner')
```

```python
# left join
pd.merge (df1, df2, on = 'player name', how = 'left')
```

```python
# right join
pd.merge (df1, df2, on = 'player name', how = 'right')
```

```python
# outer join : means union of left and right join
pd.merge (df1, df2, on = 'player name', how = 'outer')
```

```python
df3 = pd.DataFrame({
    'Player Name' : ['Rahul Dravid', 'Sachin Tendulkar', 'Virat Kohli', 'Ricky Ponting', 'M.S. Dhoni'],
    'Age' : [48, 50, 34, 42, 40],
})
df3
```

```python
df4 = pd.DataFrame({
    'Sports Person' : ['Rahul Dravid', 'Sachin Tendulkar', 'Virat Kohli', 'Hardik Pandya', 'M.S. Dhoni'],
    'salary' : [1000000, 500000, 2000000, 1500000, 2500000]
})

df4
```

```python
# Join Dataframes even when two DF have diffrant col name
#👉 Joins DataFrames based on matching column values.
# It checks:
# Player Name == Sports Person
pd.merge (df3, df4, left_on = 'Player Name', right_on = 'Sports Person', how = 'inner')
```

```python
pd.concat ([df3, df4], ignore_index=True)
```

```python
# its not check eny same values in a column direcly joins
# axis = 0 also gives a nan values but axis = 1 are gives not
pd.concat([df3, df4], axis=0)
```

```python
pd.concat([df3, df4], axis=1)
```
