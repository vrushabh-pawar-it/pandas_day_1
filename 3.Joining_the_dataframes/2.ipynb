```python
import numpy as np
```

```python
df = pd.DataFrame(data = np.random.randint(0, 100, (4,3)),
columns=('test1', 'test2', 'test3'),
index=['vinay', 'tushar', 'vrushabh', 'kartik'])
print(df)
```

```python
# if marks are greater than 70 or 80 replace those marks with A+ grade
df.where(df<=70, 'A+')
```

```python
# for > 30 bad marks
df.where(df['test3'] > 30, 'bad')
```

```python
# df['test3'] = df['test3'].where(df['test3'] > 30, 'bad')  
# not give a output insted of number 
# we already wrote a string 'bad' in a above program
```

```python
df.where(df != 87, 'medium') 
```

```python
# using lambda
df.where(lambda x: df != 13 ,'medium')
```
