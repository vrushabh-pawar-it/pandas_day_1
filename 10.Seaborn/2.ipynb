```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
print("the lenght of iris dataset is =" ,len(iris))
```

```python
#pairplot
# hue → tells Seaborn which column to use for grouping/coloring
# palette → tells Seaborn which colors to use for those groups.
sns.pairplot(iris, hue='species', palette='husl' )
```

```python
sns.pairplot(iris, hue = 'species', palette='husl', markers=['o','d','s'])
```

```python
df = iris.iloc[:10, :4]
df
```

```python
#heatmap
# cmap = Color Map
# annot means annotation.
# Annotation = write the actual values inside the boxes#
sns.heatmap(df, annot=True, cmap='BrBG')
```

```python
# catplot
# Draw a violin plot. kind = violin
sns.catplot(x = 'species', y = 'sepal_width', data = iris, kind = 'violin')
```
