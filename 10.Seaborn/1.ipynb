```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
arr=mpg['cylinders'].unique()
arr
```

```python
tup = tuple(mpg['cylinders'])
tup
```

```python
# displot
sns.displot(tup, bins = 6)
```

```python
sns.displot(tup, kde=True, bins = 6)
```

```python
""" 4 4 4 6 6 8 8 8

Bins:

3-4 |        Count = 0
4-5 | ███    Count = 3
5-6 |        Count = 0
6-7 | ██     Count = 2
7-8 |        Count = 0
8-9 | ███    Count = 3
Why use range(3,10,1)?
range(3,10,1)

creates:

[3, 4, 5, 6, 7, 8, 9] 
Start = 3
Stop before 10
Move by 1 each time """
```

```python
sns.displot(tup, kde=False, bins=range(3,10,1))
```

```python
mpg['origin'].unique()
```

```python
#countplot
# hue "Break each category into smaller colored groups."
sns.countplot(x='cylinders', data=mpg, color = 'red', hue = 'origin')
```

```python
# box plot
sns.boxplot(x = 'origin', y = 'acceleration', data=mpg)
```

```python
sns.jointplot(x='horsepower', y = 'acceleration',  data = mpg)
```

```python
# scatterplot
sns.scatterplot(x = 'acceleration', y = 'horsepower', data=mpg)
```

```python
sns.scatterplot(x='horsepower', y = 'acceleration', data = mpg, hue = 'origin')
```

```python
#lmplot
sns.lmplot(x = 'horsepower', y = 'acceleration', data = mpg, hue = 'origin')
```

```python
#1 subplot
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.scatterplot(x = 'acceleration', y = 'horsepower', data = mpg, )

#2 subplot
plt.subplot(2,2,2) # matplotlib line 
sns.scatterplot(x = 'acceleration', y = 'origin', data = mpg) # seaborn line

#3 subplot
plt.subplot(2,2,3) # here we used a matplotlib coz seaborn not have a subplot function
sns.scatterplot(x = 'acceleration', y ='horsepower', data = mpg)
```
