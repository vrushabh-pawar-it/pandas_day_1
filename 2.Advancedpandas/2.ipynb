```python
import pandas as pd
```

```python
tpl = ((1001, 'vrushabh'),(1002, 'aniket'), (1003, 'ajay'), (1004, 'om'), (1005, 'tushar'))
tpl
```

```python
df = pd.DataFrame(tpl, columns=['col1', 'col2'])
df
```

```python
# rename the column name
df.rename(columns={'col1': 'empid', 'col2': 'name'}, inplace=True)
df
```

```python
# add third col3 or new col
df['dept'] = ['IT', 'finance', 'engineering', 'medical', 'art']
df
```

```python
#  how to insert new col in betw existing col
df.insert(loc=1, column = 'salary2', value = [35000, 22000, 50000, 54001,4000])
```

```python
df
```

```python
#  insert phone number col in betw name and dept col
df.insert(loc=4, column = 'phone no', value=[1010101010, 84848484,487484884,48609876,57754884])
df
```

```python
df.drop("salary2", axis=1, inplace=True)
```

```python
df
```

```python
# creating a new column from existing colum
df['tax'] = df['salary1'] * 10/100
df
```

```python
#  create a new col from exiting col using lambda
df1 = df.assign(pf = lambda x: x.salary1 * 10/10)
df1
```

```python
# cal net salary net sal = salary - tax - pf
df2 = df1.assign(netsal = lambda x: x.salary1-x.tax-x.pf)
df2
```

```python
#  cal deduction = tax + pf
df3 = df1.assign(deduction = lambda x: x.tax+x.pf)
df3
```
