```python
import pandas as pd
```

```python
mydict = {'col1':[1,2,3,4,5],
          'col2': [6,7,8,9,10],
          'col3': ['one','two','three','four','five']}
mydict
```

```python
df = pd.DataFrame(mydict)
df
```

```python
df.columns
```

```python
df.col2
```

```python
df1 = df[['col1', 'col2']]
df1
```

```python
# Display all rows where:
# col1 > 1
# and col2 != 7

df[(df.col1 > 1) & (df.col2 != 7)]
```

```python
# dipaly col3 is not equal to three
df[(df.col3 != 'three')]
```

```python
df.iloc[0:3]
```

```python
df.iloc[0:3][['col1','col2']]
```

```python
# find the col1 values greater than 2
df.loc[df['col1']> 2]
```

```python
df.loc[0:3].query('col1> 2')
```

```python
df.query('2*col1 >= col2')
```

```python
# if we use single quote in brackets then use outer bracket double quote ["['']"]
df.query("col3 in['three','four']")
```

```python
df.query("col3 not in['three','four']")
```

```python
df.query("col2 != [5,7]")
```
